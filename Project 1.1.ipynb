```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import pandas as pd 

#list of strings
lst = ["Age", "Year", "Auxillary nodes", "Survival"]
df= pd.DataFrame(lst)

print(df)

data = pd.read_csv("cancer.csv")
                   
data
```

```python
data.describe()
```

```python
data.plot.box()
```

```python
# Max = 75% + 1.5* (75% - 25%)
maximum = 4 + 1.5* (4 - 0)

%matplotlib notebook

new_data = data[data["Auxillary nodes"] <= maximum]
new_data.plot.box()
```

```python
corr = new_data.corr()
corr
```

There is a negative correlation between Age and Auxillary nodes and also between Year and Auxillary nodes.

```python
%matplotlib inline

sns.heatmap(corr)
plt.show()
```
